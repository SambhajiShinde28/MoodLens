```python
import pandas as pd
```

```python
data = pd.read_csv("/content/Sentiment-Dataset.csv",encoding='utf-8')
```

```python
print("Basic Info:")
print(data.info())

print(f"\nDataset Shape: {data.shape}")  
```

```python
print("\nSample Data:")
print(data.head())
```

```python
print("\nSummary Statistics:")
print(data.describe(include="all"))
```

```python
print("\nMissing Values:")
print(data.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
print("\nSentiment Counts:")
print(data['sentiment'].value_counts())
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(data=data, x='sentiment', palette="coolwarm")
plt.title("Sentiment Distribution")
plt.xlabel("Sentiment")
plt.ylabel("Count")
plt.show()
```

```python
data['review_length'] = data['review'].apply(len)

plt.figure(figsize=(10, 6))
sns.histplot(data['review_length'], bins=30, kde=True, color="blue")
plt.title("Review Length Distribution")
plt.xlabel("Review Length")
plt.ylabel("Frequency")
plt.show()
```

```python
from collections import Counter
from wordcloud import WordCloud

all_reviews = " ".join(data['review'])

word_freq = Counter(all_reviews.split())

print("\nMost Common Words:")
print(word_freq.most_common(10))

wordcloud = WordCloud(width=800, height=400, background_color="white").generate(all_reviews)

plt.figure(figsize=(12, 8))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Reviews")
plt.show()
```
